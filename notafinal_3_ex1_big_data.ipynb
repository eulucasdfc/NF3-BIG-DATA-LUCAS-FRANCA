```python
#1. Obtenha o dataset do endereço https://www.w3resource.com/python-exercises/pandas/filter/world_alcohol.csv e realize as seguintes análises (2,5 pontos):
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/eulucasdfc/NF3-BIG-DATA-LUCAS-FRANCA/main/world_alcohol.csv', sep = ',')
df
```

```python
#a. Agrupe os dados por tipo de bebidas;
import pandas as pd

agrupar_dados_bebida = df.groupby('Beverage Types')['Display Value'].std()
agrupar_dados_bebida
```

```python
#b. Agrupe os dados por Região e por Ano;
import pandas as pd

agrupar_dados_rg_ano = df.groupby(['WHO region','Year']).std()
agrupar_dados_rg_ano
```

```python
#c. Seção de Contagens: Contar a ocorrência de Regiões, de Países e a soma da coluna de valores por Bebida.
import pandas as pd
#soma da coluna de valores por bebida
soma_bebida = df.groupby('Beverage Types')['Display Value'].sum()
print("Soma bebidas \n")
print(soma_bebida)

#ocorrencia regiao
ocorrencia1 = df['WHO region'].value_counts()
print("\n Ocorrencia por Região\n ")
print(ocorrencia1)

#ocorrencia regiao
ocorrencia2 = df['Country'].value_counts()
print("\n Ocorrencia por País\n ")
print(ocorrencia2)
```

```python
#d. Realize análises estatísticas da coluna dos valores: Média, Moda, Mediana, Estatística Descritiva e
#Gráfico de comparação dos valores agrupados por tipo de bebida.

print("Média: ",df["Display Value"].mean())
print("\nModa: ", df["Display Value"].mode())
print("\nMediana: ", df["Display Value"].median())
print("\nEstatística Descritiva\n", df["Display Value"].describe())

groupy_bebidas_value = df.groupby('Beverage Types')['Display Value'].std()
groupy_bebidas_value.plot.bar(x='App', y='Dislay Value')
```

```python
#e. Mostre resultados de acordo com alguns critérios:
#i. Mostrar a coluna de bebidas do ano de 1985.
import pandas as pd
df_auxiliar = df.drop(columns = ['WHO region', 'Country', 'Display Value'])
bebida = df_auxiliar.loc[df['Year'] == 1985]
bebida
```

```python
#ii. Mostrar a coluna de Região com valores acima de 4
df_auxiliar2 = df.drop(columns = ['Beverage Types', 'Country', 'Year'])
regiao = df_auxiliar2.loc[df['Display Value'] > 4]
regiao
```
