```python
#2. Seja o dataset chamado cursos-prouni.csv que se encontra no endereço Cursos e notas de corte do
#PROUNI 2018 - Datasets - Brasil.IO. Baixe o arquivo, armazene no seu github** e obtenha os dados a
#partir desse endereço do github e realize as seguintes análises (2,5 pontos):
import pandas as pd

prouni = pd.read_csv('https://raw.githubusercontent.com/eulucasdfc/NF3-BIG-DATA-LUCAS-FRANCA/main/cursos-prouni.csv', sep = ',')
prouni
```

```python
#a. Efetuar a limpeza das colunas de notas: onde tiver NaN (Not a Number), substituir por 0,0.
import pandas as pd
prouni_fillna = prouni.fillna(0)

prouni_fillna
```

```python
#b. Agrupe os dados pelo grau (Bacharelado, Licenciatura, etc).
import pandas as pd
groupby_grau = prouni.groupby('grau')

groupby_grau.std()
```

```python
#c. Agrupe os dados pelos cursos de Matemática, Medicina e Pedagogia.
import pandas as pd
cursos = pd.read_csv('https://raw.githubusercontent.com/eulucasdfc/NF3-BIG-DATA-LUCAS-FRANCA/main/cursos-prouni.csv', sep = ',', index_col = 'curso_busca')

groupby_cursos = cursos.loc[["Matemática", "Medicina", "Pedagogia"]].groupby('curso_busca')
groupby_cursos.std()
```

```python
#d. Agrupe os dados por Estado e obtenha a média de notas de corte por Estado.
import pandas as pd

groupby_estado = prouni.groupby('uf_busca')['nota_integral_ampla'].mean()
print("Média das notas por Estado\n", groupby_estado)
```

```python
#e. Agrupe os dados pelos cursos Tecnológicos.
import pandas as pd
grau = pd.read_csv('https://raw.githubusercontent.com/eulucasdfc/NF3-BIG-DATA-LUCAS-FRANCA/main/cursos-prouni.csv', sep = ',', index_col = 'grau')

groupby_tecno = grau.loc[["Tecnológico"]].groupby('curso_busca')
groupby_tecno.std()
```

```python
#f. Elimine a coluna “cidade_filtro” do dataframe.
import pandas as pd
prouni.drop(columns=['cidade_filtro'])
```

```python
#g. Apresente a média das mensalidades dos cursos de Medicina.
import pandas as pd
medicina = cursos.loc[["Medicina"]]

print("Média: ", medicina['mensalidade'].mean())
```

```python
#h. Média das notas de corte dos cursos de tempo integral.
import pandas as pd

tempo_nota = prouni[(prouni['turno'] == 'Integral') & (prouni['nota_integral_ampla'])]

print("Média: ", tempo_nota['nota_integral_ampla'].mean())
```

```python
#i. Estatística Descritiva das Notas Integral Ampla dos cursos de Bacharelado.
import pandas as pd

bacharel_nota = prouni[(prouni['grau'] == 'Bacharelado') & (prouni['nota_integral_ampla'])]

notas = bacharel_nota['nota_integral_ampla'].describe()

print("Estatistica Descritiva\n", notas)
```

```python
#j. Gráfico comparativo entre o grau dos cursos (Bacharelado, Licenciatura, Tecnologia, etc) pelas Notas Integral de Cotas.

groupby_grau_nota = prouni.groupby('grau')['nota_integral_ampla'].std()
groupby_grau_nota

groupby_grau_nota.plot.bar(x='App', y='nota_integral_ampla')
```
